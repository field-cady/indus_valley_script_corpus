```python
import os, json
import pandas as pd
```

```python
dr1 = '/home/taco/Desktop/indus_valley_script_corpus/corpus/m001_m099/'
dr2 = '/home/taco/Desktop/indus_valley_script_corpus/corpus/m100_m199/'

blobs = [json.loads(open(dr1+f).read()) for f in sorted(os.listdir(dr1))]
blobs += [json.loads(open(dr2+f).read()) for f in sorted(os.listdir(dr2))] 
```

```python
seqs = []
for b in blobs: seqs.append([x['id'] for x in b[0]['graphemes']])

seqs
```

```python
df = pd.DataFrame([b[0] for b in blobs])
df['length'] = df.graphemes.apply(len)
df['seq'] = df.graphemes.apply(lambda lst: tuple(g['id'] for g in lst))
df['fst'] = df.seq.apply(lambda s: s[0])
df['lst'] = df.seq.apply(lambda s: s[-1])

df
```
